```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lifelines import CoxPHFitter

# 생성할 데이터의 수
n = 100

# 각 그룹의 데이터를 생성
time_no_statin = np.random.exponential(scale=10, size=n)
time_low_intensity = np.random.exponential(scale=12, size=n)
time_moderate_intensity = np.random.exponential(scale=14, size=n)
time_high_intensity = np.random.exponential(scale=16, size=n)

# 이벤트 발생 여부 (0: 검열됨, 1: 이벤트 발생)
event_no_statin = np.random.binomial(1, 0.8, size=n)
event_low_intensity = np.random.binomial(1, 0.7, size=n)
event_moderate_intensity = np.random.binomial(1, 0.6, size=n)
event_high_intensity = np.random.binomial(1, 0.5, size=n)

# 데이터 프레임 생성
data = pd.DataFrame({
    'time': np.concatenate([time_no_statin, time_low_intensity, time_moderate_intensity, time_high_intensity]),
    'event': np.concatenate([event_no_statin, event_low_intensity, event_moderate_intensity, event_high_intensity]),
    'group': ['No statin']*n + ['Low Intensity']*n + ['Moderate Intensity']*n + ['High Intensity']*n
})

# 더미 변수 생성
data = pd.get_dummies(data, columns=['group'], drop_first=True)

# CoxPHFitter 모델 적합
cph = CoxPHFitter()
cph.fit(data, duration_col='time', event_col='event')

# 요약 출력
cph.print_summary()

# 계수 플롯
cph.plot()
plt.title('Cox Proportional Hazards Model Coefficients')
plt.savefig('cox.jpg',dpi=300)
plt.show()
```

```python
import pandas as pd
import numpy as np
from lifelines import CoxPHFitter

# Generating hypothetical data
np.random.seed(42)
n = 500  # number of samples per group

# Creating survival times and event occurrences for each group
time_no_statin = np.random.exponential(scale=200, size=n)
event_no_statin = np.random.binomial(1, 0.5, size=n)

time_low_intensity = np.random.exponential(scale=220, size=n)
event_low_intensity = np.random.binomial(1, 0.5, size=n)

time_moderate_intensity = np.random.exponential(scale=240, size=n)
event_moderate_intensity = np.random.binomial(1, 0.5, size=n)

time_high_moderate_intensity = np.random.exponential(scale=260, size=n)
event_high_moderate_intensity = np.random.binomial(1, 0.5, size=n)

time_high_intensity = np.random.exponential(scale=280, size=n)
event_high_intensity = np.random.binomial(1, 0.5, size=n)

# Combining data into a single DataFrame
df = pd.DataFrame({
    'time': np.concatenate([time_no_statin, time_low_intensity, time_moderate_intensity, time_high_moderate_intensity, time_high_intensity]),
    'event': np.concatenate([event_no_statin, event_low_intensity, event_moderate_intensity, event_high_moderate_intensity, event_high_intensity]),
    'group': np.concatenate([np.repeat('no_statin', n), np.repeat('low_intensity', n), np.repeat('moderate_intensity', n), np.repeat('high_moderate_intensity', n), np.repeat('high_intensity', n)])
})

# Creating dummy variables for group
df = pd.get_dummies(df, columns=['group'], drop_first=True)

# Cox proportional hazards model
cph = CoxPHFitter()
cph.fit(df, duration_col='time', event_col='event')

# Print the summary
print(cph.summary)
```

```python


# 계수 플롯
cph.plot()
plt.title('Cox Proportional Hazards Model Coefficients')
plt.savefig('cox.jpg',dpi=300)
plt.show()
```
